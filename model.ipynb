Read video and return array of frames 

Frames are downsized to 256 x 256

```python
import cv2
import numpy as np
from PIL import Image

def readVideo(filename):
    # Create a VideoCapture object and read from input file
    # If the input is the camera, pass 0 instead of the video file name
    cap = cv2.VideoCapture(filename)
    video = []

    # Check if camera opened successfully
    if (cap.isOpened()== False): 
      print("Error opening video stream or file")

    # Read until video is completed
    for i in range(1000): ret, frame = cap.read()
    while(cap.isOpened()):
      # Capture frame-by-frame
      for i in range(60): ret, frame = cap.read()
      if ret == True:

        # Display the resulting frame
        im = Image.fromarray(frame).resize((256,256))
        video.append(np.asarray(im))

        # Press Q on keyboard to  exit
        if cv2.waitKey(25) & 0xFF == ord('q'):
          break

      # Break the loop
      else: 
        break

    # When everything done, release the video capture object
    cap.release()

    # Closes all the frames
    cv2.destroyAllWindows()

    return video
```

Assign videos to a data variable

```python
# import required module
import os
import pandas as pd
# assign directory
directory = "/content/drive/MyDrive/fake news challenge/data"
data = pd.read_csv('/content/drive/MyDrive/fake news challenge/train.csv')

x = []
y = []
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        x.append(readVideo(f))
        y.append(data['label'][data['filename'].tolist().index(filename)])
```
